**Number of matches won by each team in ipl 2022**

```python
figure=px.bar(data_frame=data,x="match_winner",title="Number of Matches won by each team")
```

```python
figure.show()
```

```python
data["won_by"]=data["won_by"].map({'Wickets':'Chasing','Runs':'Defending'})
won_by=data["won_by"].value_counts()
labels=won_by.index
counts=won_by.values
colors=['red','green']
fig = go.Figure(data=[go.Pie(labels=labels, values=counts)])
fig.update_layout(title="Number Of Matches won by defender or chasser")
fig.show()
```

```python
figure=px.bar(data,x="best_bowling",title="Best Bowler in each match in IPL 2022")
figure.show()
```

```python
figure=px.bar(data,x="player_of_the_match",title="Player of the match in IPL 2022 of every match")
figure.show()
```
